```python
# Copying the column Authors_Journal into a new list
Journals_A= df['Authors_Journal'].copy()

Journals_A.head()
```

```python
#Seperating the Authors and Journals
Authors=[]
Journals=[]

for word in Journals_A:
    before= word.split("–")[0]
    Authors.append(before)
    after = word.split('–', 1)[-1].strip()
    Journals.append(after)
```

```python
# Adding the seperated columns of Authors and Journals into the dataframe 'df'
df['Authors']=Authors
df['Journals']=Journals
```

```python
df. head()
```

```python
# Deleting the unnecessary column
df.drop('Authors_Journal',axis=1,inplace=True)
```

```python
# Creating a new csv file for the clean data
df.to_csv(r"C:\Users\admin\project_DB_design\articles_cleaned.csv")
```
